```python
import matplotlib.pyplot as plt
import numpy as np

# Collect all normalized dB values
all_db_values = []
for i, (batch_data, _) in enumerate(test_loader):
    all_db_values.extend(batch_data.numpy().flatten())
    if i >= 1:  # Limit for memory
        break

all_db_values = np.array(all_db_values)
plt.hist(all_db_values, bins=100, color='purple', alpha=0.7)
plt.xlabel('Normalized dB Magnitude')
plt.ylabel('Frequency')
plt.title('Histogram of Normalized dB Magnitude')
plt.show()
```
